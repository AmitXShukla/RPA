- Jupyter notebook access from local network
To access a Jupyter Notebook running on your local machine from another device on your local network, you can follow these steps:

    Start a Jupyter Notebook server on your local machine by opening a terminal or command prompt and navigating to the directory containing your notebooks. Then, run the command jupyter notebook. This should start the server and open the Jupyter Notebook interface in your default web browser.

    Once the Jupyter Notebook interface is open, look for the URL in the address bar of your browser. The URL should be something like <http://localhost:8888/>. Take note of the port number (in this example, 8888).

    On the device you want to access the Jupyter Notebook from, open a web browser and enter the IP address of your local machine followed by a colon and the port number. For example, if your local machine has an IP address of 192.168.1.2 and the port number is 8888, you would enter <http://192.168.1.2:8888/> in the address bar of your web browser.

    If everything is set up correctly, you should see the Jupyter Notebook interface in your web browser on the other device, and you can open and edit your notebooks just as you would on your local machine.

Note that you may need to configure your firewall or router settings to allow incoming connections on the port used by Jupyter Notebook (in the example above, port 8888). Additionally, if your local IP address changes frequently, you may need to update the URL on the other device each time the IP address changes.

- run vscode from browser
You can set up and access VSCode from your web browser by using the "code-server" application. Here are the steps to get started:

    Install "code-server" by following the instructions in their documentation: <https://github.com/cdr/code-server#installation>

    After installation, run "code-server" by typing code-server in the terminal. This will start the application and provide you with a URL that you can use to access VSCode in your browser.

    Open your web browser and navigate to the URL provided by "code-server". This will open VSCode in your browser, allowing you to edit and run code just as you would in the desktop application.

Note that when you run "code-server", it will create a default password that you will need to enter in order to access VSCode in your browser. You can also set up authentication using external providers like Google or GitHub, or by setting up your own authentication system using OAuth2.

You can also configure "code-server" to run as a service, so that it starts automatically when your server or computer boots up. The documentation provides detailed instructions on how to do this.

Additionally, you may experience some limitations when using VSCode in your browser compared to the desktop application, such as reduced performance or the inability to access certain extensions. However, "code-server" provides a useful alternative for when you need to access VSCode remotely or from a device without the desktop application installed.

- Python HTTP server
To run and access a Python HTTP server from your local network, you can follow these steps:

    Open a command prompt or terminal window on your local machine and navigate to the directory where you want to serve files from.

    Start a Python HTTP server by running the command python -m http.server (Python 3) or python -m SimpleHTTPServer (Python 2). This will start the server and begin serving files from the current directory.

    By default, the server will listen on port 8000. To access the server from another device on your local network, you need to find the IP address of your local machine. You can do this by running the command ipconfig on Windows, or ifconfig on Linux or macOS. Look for the IPv4 address of your machine, which should be in the format of xxx.xxx.xxx.xxx.

    On the device you want to access the Python HTTP server from, open a web browser and enter the IP address of your local machine followed by a colon and the port number used by the server (in this example, 8000). For example, if your local machine has an IP address of 192.168.1.2 and the server is running on port 8000, you would enter <http://192.168.1.2:8000/> in the address bar of your web browser.

    If everything is set up correctly, you should see the files being served by the Python HTTP server in your web browser on the other device.

Note that this is a simple way to serve files over HTTP from your local machine and is not intended for production use. If you need more advanced features or want to host a web application, you should consider using a dedicated web server or a cloud hosting service.
