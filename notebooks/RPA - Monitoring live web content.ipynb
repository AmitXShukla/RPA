## Live Automation project to monitor web content

```python
from lxml import html
import requests
import os
import winsound

def checkStatus():
    page = requests.get('https://www.heliyatra.irctc.co.in/')
    tree = html.fromstring(page.content)
    #This will create a list of buyers:
    ticketStatus = tree.xpath('/html/body/div[1]/main/section/div/div[2]/p[1]/text()')
    if ticketStatus[0] != "Ticket Booking Opening Soon":
        # play sound or send email
        # below code just open an image file
        os.system("start C:\\amit.la\\WIP\YT\\banner17.png")
    print('status: ', ticketStatus)
```

```python
import time

# Set the interval for checking status
interval = 300  # in seconds
counter = 0

while counter < 1000:
    checkStatus()
    counter +=1 
    time.sleep(interval)
    print(counter)
```
