## code Generated by ChatGPT

```python
import os
from cryptography.fernet import Fernet

def encrypt_file(file_path, key):
    """
    Encrypts a file using AES encryption algorithm and saves the encrypted data
    to a new file with the .enc extension.
    :param file_path: the path of the file to be encrypted
    :param key: the encryption key to use (must be 32 bytes long)
    """
    # Generate a Fernet key from the provided key
    fernet_key = Fernet(key)
    # Read the contents of the file to be encrypted
    with open(file_path, "rb") as f:
        data = f.read()
    # Encrypt the data using the Fernet key
    encrypted_data = fernet_key.encrypt(data)
    # Save the encrypted data to a new file with the .enc extension
    encrypted_file_path = os.path.splitext(file_path)[0] + ".enc"
    with open(encrypted_file_path, "wb") as f:
        f.write(encrypted_data)

def decrypt_file(file_path, key):
    """
    Decrypts a file that has been encrypted using AES encryption algorithm and
    saves the decrypted data to a new file with the .dec extension.
    :param file_path: the path of the file to be decrypted
    :param key: the decryption key to use (must be 32 bytes long)
    """
    # Generate a Fernet key from the provided key
    fernet_key = Fernet(key)
    # Read the contents of the encrypted file
    with open(file_path, "rb") as f:
        encrypted_data = f.read()
    # Decrypt the data using the Fernet key
    decrypted_data = fernet_key.decrypt(encrypted_data)
    # Save the decrypted data to a new file with the .dec extension
    decrypted_file_path = os.path.splitext(file_path)[0] + ".dec"
    with open(decrypted_file_path, "wb") as f:
        f.write(decrypted_data)
```

```python
# Generate a random 32-byte encryption key
key = Fernet.generate_key()
```

```python
# Encrypt a file using the key
encrypt_file("../SampleData/sampleData.csv", key)
```

```python
# Decrypt the encrypted file using the same key
decrypt_file("../SampleData/sampleData.csv.enc", key)
```

This would encrypt the plaintext.txt file using the generated key and save the encrypted data to a new file called plaintext.txt.enc. It would then decrypt the encrypted file and save the decrypted data to a new file called plaintext.txt.dec. Note that the original plaintext.txt file is not modified by these functions.
